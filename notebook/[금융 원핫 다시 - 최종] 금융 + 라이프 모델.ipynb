```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder, LabelEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.ensemble import RandomForestClassifier, VotingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import accuracy_score, classification_report, roc_auc_score, confusion_matrix
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
import matplotlib.pyplot as plt
import seaborn as sns

# ë°ì´í„° ë¶ˆëŸ¬ì˜¤ê¸°
df = pd.read_csv('../data/df_KCB_grade.csv')
target_col = 'KCB_grade'
```

```python
#  1. ë³€ìˆ˜ ì •ì˜
onehot_features = ['SEX', 'JB_TP', 'HOME_ADM', 'COM_ADM', 'HB_1ST', 'B1Y_EQP_MDL', 'B1Y_MOB_OS','FAM_OWN_LIV_YN','OWN_LIV_YN','AGE','ROP']
label_features = ['LIF_STG']
binary_features = ['VIP_CARD_YN', 'CAR_YN', 'TRAVEL_OS', 'TRAVEL_JJ', 'GOLF_INDOOR',
                   'PREFER_SPORTS', 'RNCAR_JJ_YN', 'AIRLINE_OS_YN']
continuous_features = [
    'R3M_FOOD_AMT', 'R3M_ENT_AMT', 'R3M_DEP_AMT', 'R3M_TRAVEL_AMT',
    'R3M_BEAUTY_AMT', 'R3M_ITRT_FIN_PAY', 'R3M_ITRT_FIN_INSUR',
    'R3M_ITRT_FIN_BANK', 'R3M_ITRT_LIFE_HEALTH', 'R3M_ITRT_SHOP_MART',
    'R3M_ITRT_ENT_SVOD'
]
fin_col = [
    'TOT_ASST', 'HOUS_LN_BAL', 'CRDT_LN_BAL', 'OWN_HOUS_CNT', 
    'FAM_OWN_HOUS_CNT', 'CRDT_LN_BAL_NEW',
    'PYE_MAX_DLQ_DAY', 'PYE_C1L120237', 'PYE_C1L120250',
    'PYE_L10210000', 'PYE_L10216000', 'PYE_L10231000',
    'CD_USE_AMT', 'YR_SALES', 'SHP_CD_USE', 'SHP_LN_BAL',
    'PYE_D10110001', 'PYE_D10110003', 'PYE_D10110006',
    'PYE_D1011000C', 'PYE_D10210D00', 'PYE_D10232000'
]



#  2. ë¼ë²¨ ì¸ì½”ë”©
df = df.copy()
label_encoder = LabelEncoder()
df['LIF_STG'] = label_encoder.fit_transform(df['LIF_STG'])


#  3. ì „ì²˜ë¦¬ ì •ì˜
preprocessor = ColumnTransformer([
    ('onehot', OneHotEncoder(drop='first', handle_unknown='ignore'), onehot_features),
    ('scaler', StandardScaler(), binary_features + continuous_features + fin_col)
])

#  4. ì „ì²˜ë¦¬ ì ìš©
processed_array = preprocessor.fit_transform(df)
onehot_names = preprocessor.named_transformers_['onehot'].get_feature_names_out(onehot_features)
scaled_names = binary_features + continuous_features + fin_col
all_columns = np.concatenate([onehot_names, scaled_names])
processed_df = pd.DataFrame(processed_array, columns=all_columns)
processed_df['LIF_STG'] = df['LIF_STG']

X = processed_df
y = df['KCB_grade']

#  5. í•™ìŠµ/í…ŒìŠ¤íŠ¸ ë¶„ë¦¬
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)



#  6. í‰ê°€ í•¨ìˆ˜ ì •ì˜
def grid_search_rf(X, y):
    params = {'n_estimators': [100], 'max_depth': [30, 50, 100], 'min_samples_split': [2, 3, 4, 5]}
    clf = GridSearchCV(RandomForestClassifier(random_state=42), params, cv=3, scoring='roc_auc', n_jobs=-1)
    clf.fit(X, y)
    print(" RF best:", clf.best_params_)
    return clf.best_estimator_

def grid_search_xgb(X, y):
    params = {'n_estimators': [100], 'max_depth': [30, 50, 100], 'learning_rate': [0.1, 0.2, 0.3, 0.4]}
    clf = GridSearchCV(XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42),
                       params, cv=3, scoring='roc_auc', n_jobs=-1)
    clf.fit(X, y)
    print(" XGB best:", clf.best_params_)
    return clf.best_estimator_

def grid_search_lgbm(X, y):
    params = {'n_estimators': [100], 'max_depth': [30, 50, 100], 'learning_rate': [0.1, 0.2, 0.3, 0.4]}
    clf = GridSearchCV(LGBMClassifier(), params, cv=3, scoring='roc_auc', n_jobs=-1)
    clf.fit(X, y)
    print(" LGBM best:", clf.best_params_)
    return clf.best_estimator_

def print_full_classification_report(y_true, y_pred, model_name):
    report_df = pd.DataFrame(classification_report(y_true, y_pred, output_dict=True)).transpose()
    print(f"\nðŸ“‹ {model_name} - Classification Report")
    print(report_df.round(4))

def plot_confusion_matrix(y_test, y_pred, model_name, acc, auc):
    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(5, 4))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=["0", "1"], yticklabels=["0", "1"])
    plt.title(f"{model_name}\nConfusion Matrix\n(Accuracy: {acc:.2f}, AUC: {auc:.2f})")
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.tight_layout()
    plt.show()

def train_and_evaluate(model, X_train, X_test, y_train, y_test, name="Model"):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    y_prob = model.predict_proba(X_test)[:, 1]
    acc = accuracy_score(y_test, y_pred)
    auc = roc_auc_score(y_test, y_prob)
    print(f"\nðŸ”¹ {name} - Accuracy: {acc:.4f}, AUC: {auc:.4f}")
    print_full_classification_report(y_test, y_pred, name)
    plot_confusion_matrix(y_test, y_pred, name, acc, auc)
    return auc, y_prob

def evaluate_overfitting(model, X_train, X_test, y_train, y_test, name):
    train_auc = roc_auc_score(y_train, model.predict_proba(X_train)[:, 1])
    test_auc = roc_auc_score(y_test, model.predict_proba(X_test)[:, 1])
    train_r = classification_report(y_train, model.predict(X_train), output_dict=True)
    test_r = classification_report(y_test, model.predict(X_test), output_dict=True)
    p_diff = train_r['weighted avg']['precision'] - test_r['weighted avg']['precision']
    r_diff = train_r['weighted avg']['recall'] - test_r['weighted avg']['recall']
    f_diff = train_r['weighted avg']['f1-score'] - test_r['weighted avg']['f1-score']
    overfit = train_auc - test_auc
    label = "ë‚®ìŒ" if overfit < 0.02 and p_diff < 0.05 and r_diff < 0.05 and f_diff < 0.05 else \
            "ì¤‘ê°„ (ì£¼ì˜ í•„ìš”)" if overfit < 0.05 or p_diff < 0.07 or r_diff < 0.07 or f_diff < 0.07 else "ë†’ìŒ (ì‹¬ê°í•œ ê³¼ì í•©)"
    print(f"\nðŸ” {name} ê³¼ì í•© í‰ê°€\nAUC(train): {train_auc:.4f}, AUC(test): {test_auc:.4f}")
    print(f"Precision Î”: {p_diff:.4f}, Recall Î”: {r_diff:.4f}, F1 Î”: {f_diff:.4f} â†’ ê³¼ì í•© ìˆ˜ì¤€: {label}")


# 7. ëª¨ë¸ í•™ìŠµ ë° í‰ê°€
rf = grid_search_rf(X_train, y_train)
xgb = grid_search_xgb(X_train, y_train)
lgbm = grid_search_lgbm(X_train, y_train)
logreg = LogisticRegression(max_iter=1000)
logreg.fit(X_train, y_train)

models = [rf, xgb, lgbm, logreg]
model_names = ["RF", "XGB", "LGBM", "LogReg"]


for model, name in zip(models, model_names):
    _, _ = train_and_evaluate(model, X_train, X_test, y_train, y_test, name)
    evaluate_overfitting(model, X_train, X_test, y_train, y_test, name)

#  8. ì•™ìƒë¸”
ensemble = VotingClassifier(estimators=[
    ('rf', rf), ('xgb', xgb), ('lgbm', lgbm), ('logreg', logreg)
], voting='soft')
ensemble.fit(X_train, y_train)

train_and_evaluate(ensemble, X_train, X_test, y_train, y_test, "Ensemble")
evaluate_overfitting(ensemble, X_train, X_test, y_train, y_test, "Ensemble")




```

```python
from sklearn.metrics import roc_curve, auc

# âœ… ëª¨ë¸ ë¦¬ìŠ¤íŠ¸ (ì´ë¦„ê³¼ í•¨ê»˜)
models_with_names = [
    (rf, "RandomForest"),
    (xgb, "XGBoost"),
    (lgbm, "LightGBM"),
    (logreg, "LogisticRegression"),
    (ensemble, "Ensemble")
]

# âœ… ROC Curve ì‹œê°í™”
plt.figure(figsize=(10, 7))

for model, name in models_with_names:
    y_prob = model.predict_proba(X_test)[:, 1]
    fpr, tpr, _ = roc_curve(y_test, y_prob)
    roc_auc = auc(fpr, tpr)
    plt.plot(fpr, tpr, label=f"{name} (AUC={roc_auc:.4f})")

# ê¸°ì¤€ì„  ì¶”ê°€
plt.plot([0, 1], [0, 1], 'k--', lw=1)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title(' ROC Curve - All Models')
plt.legend(loc='lower right')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder, LabelEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score, roc_curve

#  ë³€ìˆ˜ ì •ì˜
onehot_features_life = ['SEX', 'JB_TP', 'HOME_ADM', 'COM_ADM', 'HB_1ST', 'B1Y_EQP_MDL', 'B1Y_MOB_OS', 'AGE']
binary_features = ['VIP_CARD_YN', 'CAR_YN', 'TRAVEL_OS', 'TRAVEL_JJ',
                   'GOLF_INDOOR', 'PREFER_SPORTS', 'RNCAR_JJ_YN', 'AIRLINE_OS_YN']
continuous_features = [
    'R3M_FOOD_AMT', 'R3M_ENT_AMT', 'R3M_DEP_AMT', 'R3M_TRAVEL_AMT',
    'R3M_BEAUTY_AMT', 'R3M_ITRT_FIN_PAY', 'R3M_ITRT_FIN_INSUR',
    'R3M_ITRT_FIN_BANK', 'R3M_ITRT_LIFE_HEALTH', 'R3M_ITRT_SHOP_MART',
    'R3M_ITRT_ENT_SVOD'
]
onehot_features_full = onehot_features_life + ['FAM_OWN_LIV_YN', 'OWN_LIV_YN']
fin_col = [
    'TOT_ASST', 'HOUS_LN_BAL', 'CRDT_LN_BAL', 'OWN_HOUS_CNT', 
    'FAM_OWN_HOUS_CNT', 'CRDT_LN_BAL_NEW', 'ROP',
    'PYE_MAX_DLQ_DAY', 'PYE_C1L120237', 'PYE_C1L120250',
    'PYE_L10210000', 'PYE_L10216000', 'PYE_L10231000',
    'CD_USE_AMT', 'YR_SALES', 'SHP_CD_USE', 'SHP_LN_BAL',
    'PYE_D10110001', 'PYE_D10110003', 'PYE_D10110006',
    'PYE_D1011000C', 'PYE_D10210D00', 'PYE_D10232000'
]

#  ëžœë¤í¬ë ˆìŠ¤íŠ¸ ê·¸ë¦¬ë“œì„œì¹˜ í•¨ìˆ˜
def grid_search_rf(X, y):
    params = {
        'n_estimators': [100],
        'max_depth': [10, 20, None],
        'min_samples_split': [2, 5]
    }
    clf = GridSearchCV(RandomForestClassifier(random_state=42), params, cv=3, scoring='roc_auc', n_jobs=-1)
    clf.fit(X, y)
    print("âœ… RF Best Params:", clf.best_params_)
    return clf.best_estimator_

#  ì „ì²˜ë¦¬ ë° ëª¨ë¸ í•™ìŠµ í•¨ìˆ˜
def preprocess_and_train_rf(df, mode='life'):
    df = df.copy()
    df['LIF_STG'] = LabelEncoder().fit_transform(df['LIF_STG'])

    if mode == 'life':
        features = onehot_features_life
        scaler_features = binary_features + continuous_features + ['LIF_STG']

        preprocessor = ColumnTransformer([
            ('onehot', OneHotEncoder(drop='first', handle_unknown='ignore', sparse_output=False), features),
            ('scaler', StandardScaler(), scaler_features)
        ])
        X = preprocessor.fit_transform(df)
        y = df['KCB_grade']

    elif mode == 'fin':
        df_encoded = pd.get_dummies(df, columns=['FAM_OWN_LIV_YN', 'OWN_LIV_YN'], prefix=['FAM_OWN_LIV_YN', 'OWN_LIV_YN'], dtype=int)
        fin_col_filtered = [col for col in fin_col if col not in ['FAM_OWN_LIV_YN', 'OWN_LIV_YN']]
        encoded_columns = [col for col in df_encoded.columns if 'FAM_OWN_LIV_YN' in col or 'OWN_LIV_YN' in col]
        X = df_encoded[fin_col_filtered + encoded_columns]
        y = df_encoded['KCB_grade']

    elif mode == 'full':
        features = onehot_features_full
        scaler_features = binary_features + continuous_features + fin_col + ['LIF_STG']

        preprocessor = ColumnTransformer([
            ('onehot', OneHotEncoder(drop='first', handle_unknown='ignore', sparse_output=False), features),
            ('scaler', StandardScaler(), scaler_features)
        ])
        X = preprocessor.fit_transform(df)
        y = df['KCB_grade']

    # í•™ìŠµ
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
    model = grid_search_rf(X_train, y_train)
    y_prob = model.predict_proba(X_test)[:, 1]
    auc_score = roc_auc_score(y_test, y_prob)
    fpr, tpr, _ = roc_curve(y_test, y_prob)
    return fpr, tpr, auc_score

#  ì„¸ ëª¨ë¸ ì‹¤í–‰
fpr_life, tpr_life, auc_life = preprocess_and_train_rf(df, 'life')
fpr_fin, tpr_fin, auc_fin = preprocess_and_train_rf(df, 'fin')
fpr_full, tpr_full, auc_full = preprocess_and_train_rf(df, 'full')

#  ROC Curve ì‹œê°í™”
plt.figure(figsize=(10, 7))
plt.plot(fpr_life, tpr_life, label=f"LIFE ë°ì´í„° (AUC={auc_life:.2f})", color='green')
plt.plot(fpr_fin, tpr_fin, label=f"ê¸ˆìœµ ë°ì´í„° (AUC={auc_fin:.2f})", color='orange')
plt.plot(fpr_full, tpr_full, label=f"LIFE+ê¸ˆìœµ ë°ì´í„° (AUC={auc_full:.2f})", color='blue')
plt.plot([0, 1], [0, 1], 'k--')

plt.title(" ROC Curve - RandomForest (GridSearchCV ê¸°ë°˜)", fontsize=14)
plt.xlabel("False Positive Rate (FPR)")
plt.ylabel("True Positive Rate (TPR)")
plt.legend(loc='lower right')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder, LabelEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.ensemble import RandomForestClassifier, VotingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score, StratifiedKFold
from sklearn.metrics import accuracy_score, classification_report, roc_auc_score, confusion_matrix
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
import matplotlib.pyplot as plt
import seaborn as sns

# ë°ì´í„° ë¶ˆëŸ¬ì˜¤ê¸°
df = pd.read_csv('../data/df_KCB_grade.csv')
target_col = 'KCB_grade'

# 1. ë³€ìˆ˜ ì •ì˜
onehot_features = ['SEX', 'JB_TP', 'HOME_ADM', 'COM_ADM', 'HB_1ST', 'B1Y_EQP_MDL', 'B1Y_MOB_OS','FAM_OWN_LIV_YN','OWN_LIV_YN','AGE','ROP']
label_features = ['LIF_STG']
binary_features = ['VIP_CARD_YN', 'CAR_YN', 'TRAVEL_OS', 'TRAVEL_JJ', 'GOLF_INDOOR',
                   'PREFER_SPORTS', 'RNCAR_JJ_YN', 'AIRLINE_OS_YN']
continuous_features = [
    'R3M_FOOD_AMT', 'R3M_ENT_AMT', 'R3M_DEP_AMT', 'R3M_TRAVEL_AMT',
    'R3M_BEAUTY_AMT', 'R3M_ITRT_FIN_PAY', 'R3M_ITRT_FIN_INSUR',
    'R3M_ITRT_FIN_BANK', 'R3M_ITRT_LIFE_HEALTH', 'R3M_ITRT_SHOP_MART',
    'R3M_ITRT_ENT_SVOD'
]
fin_col = [
    'TOT_ASST', 'HOUS_LN_BAL', 'CRDT_LN_BAL', 'OWN_HOUS_CNT', 
    'FAM_OWN_HOUS_CNT', 'CRDT_LN_BAL_NEW',
    'PYE_MAX_DLQ_DAY', 'PYE_C1L120237', 'PYE_C1L120250',
    'PYE_L10210000', 'PYE_L10216000', 'PYE_L10231000',
    'CD_USE_AMT', 'YR_SALES', 'SHP_CD_USE', 'SHP_LN_BAL',
    'PYE_D10110001', 'PYE_D10110003', 'PYE_D10110006',
    'PYE_D1011000C', 'PYE_D10210D00', 'PYE_D10232000'
]

# 2. ë¼ë²¨ ì¸ì½”ë”©
df = df.copy()
label_encoder = LabelEncoder()
df['LIF_STG'] = label_encoder.fit_transform(df['LIF_STG'])

# 3. ì „ì²˜ë¦¬ ì •ì˜
preprocessor = ColumnTransformer([
    ('onehot', OneHotEncoder(drop='first', handle_unknown='ignore'), onehot_features),
    ('scaler', StandardScaler(), binary_features + continuous_features + fin_col)
])

# 4. ì „ì²˜ë¦¬ ì ìš©
processed_array = preprocessor.fit_transform(df)
onehot_names = preprocessor.named_transformers_['onehot'].get_feature_names_out(onehot_features)
scaled_names = binary_features + continuous_features + fin_col
all_columns = np.concatenate([onehot_names, scaled_names])
processed_df = pd.DataFrame(processed_array, columns=all_columns)
processed_df['LIF_STG'] = df['LIF_STG']

X = processed_df
y = df[target_col]

# 5. í•™ìŠµ/í…ŒìŠ¤íŠ¸ ë¶„ë¦¬
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

# 6. ëª¨ë¸ ì •ì˜ ë° í•™ìŠµ
rf = GridSearchCV(RandomForestClassifier(random_state=42),
                  {'n_estimators': [100], 'max_depth': [100, 20, None], 'min_samples_split': [2, 5]},
                  cv=3, scoring='roc_auc', n_jobs=-1).fit(X_train, y_train).best_estimator_

xgb = GridSearchCV(XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42),
                   {'n_estimators': [100], 'max_depth': [3, 6], 'learning_rate': [0.1, 0.3]},
                   cv=3, scoring='roc_auc', n_jobs=-1).fit(X_train, y_train).best_estimator_

lgbm = GridSearchCV(LGBMClassifier(random_state=42),
                    {'n_estimators': [100], 'max_depth': [10, 20], 'learning_rate': [0.05, 0.1]},
                    cv=3, scoring='roc_auc', n_jobs=-1).fit(X_train, y_train).best_estimator_

logreg = LogisticRegression(max_iter=1000, random_state=42).fit(X_train, y_train)

ensemble = VotingClassifier(estimators=[('rf', rf), ('xgb', xgb), ('lgbm', lgbm), ('logreg', logreg)], voting='soft').fit(X_train, y_train)

models = [rf, xgb, lgbm, logreg, ensemble]
names = ['RF', 'XGB', 'LGBM', 'LogReg', 'Ensemble']

# 7. êµì°¨ê²€ì¦ìœ¼ë¡œ ì¼ë°˜í™” ì„±ëŠ¥ í‰ê°€
for model, name in zip(models, names):
    skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
    scores = cross_val_score(model, X, y, cv=skf, scoring='roc_auc', n_jobs=-1)
    print(f"\nâœ… {name} êµì°¨ê²€ì¦ ê²°ê³¼ (ROC-AUC):")
    print(f" - AUC Scores: {np.round(scores,4)}")
    print(f" - AUC í‰ê· : {scores.mean():.4f}, í‘œì¤€íŽ¸ì°¨: {scores.std():.4f}")
```
